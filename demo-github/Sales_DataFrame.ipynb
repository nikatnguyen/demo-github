```python
#Dependencies and Setup
import pandas as pd
from pathlib import Path

#Files To Load
restaurant_sales_to_load = Path("restaurant_sales.csv")
vehicle_sales_to_load = Path("vehicle_sales.csv")

#Read Restaurant and Vehicle Data Files and store into Pandas DataFrames
restaurant_sales = pd.read_csv(restaurant_sales_to_load)
vehicle_sales = pd.read_csv(vehicle_sales_to_load)

#Combine the Data into a single dataset
sales_data_complete = pd.merge(restaurant_sales, vehicle_sales, how="left")
sales_data_complete
```

```python
#Rename Columns
sales_data_complete = sales_data_complete.rename(columns = {"DATE":"Year", "MRTSSM7225USN":"Total Restaurant Sales (in millions)", "TOTALSA":"Total Vehicle Sales (in millions)"})
sales_data_complete
```

```python
def extract_year(date_str):
    year = date_str.split('-')[0]
    return year

# Use a for loop to apply the function to each element in the 'Year' column
for i in range(len(sales_data_complete['Year'])):
    sales_data_complete['Year'][i] = extract_year(sales_data_complete['Year'][i])

# Display the updated DataFrame
sales_data_complete
```

```python
#Convert Year Column to show Total Sales in Years
sales_data_complete['Year'] = pd.to_datetime(sales_data_complete['Year'])
result = sales_data_complete.groupby(pd.Grouper(key='Year', freq='Y')).sum()
result

```

```python
result.index = result.index.year.astype(str)
result.reset_index(inplace=True)
```

```python
#Rewriting the 'Year' column to be just the year
def extract_year(date_str):
    year = date_str.split('-')[0]
    return year

# Use a for loop to apply the function to each element in the 'Year' column
for i in range(len(result['Year'])):
    result['Year'][i] = extract_year(result['Year'][i])

# Display the updated DataFrame
result
```

```python
#Save result dataframe to csv for analysis
result.to_csv('sales_by_year.csv')
```
