```python
#Import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
#Import csv files to read
sales_file_path = Path("sales_by_year.csv")
unemployment_file_path = Path("unemployment_by_year.csv")

#read csv files
sales_df = pd.read_csv(sales_file_path)
unemployment_df = pd.read_csv(unemployment_file_path)
```

```python
sales_df.head()
```

```python
unemployment_df.head()
```

```python
#Cleaning up the dfs
sales_reduced_df = sales_df[["Year", "Total Restaurant Sales (in millions)", "Total Vehicle Units (in millions)"]]
sales_reduced_df
```

```python
unemployment_reduced_df = unemployment_df[["Year", "Yearly Unemployment Rate"]]
unemployment_reduced_df
```

```python
#Merge the two cleaned dataframes
merged_df = pd.merge(unemployment_reduced_df, sales_reduced_df, on="Year", how="right")
merged_df
```

```python
#Drop any rows with NAN value
merged_df = merged_df.dropna(how='any')
merged_df
```

```python
#Save as new csv file to make graphs
merged_df.to_csv('unemployment_and_sales.csv')
```

### Mj Graphs

```python
nojobyear = merged_df["Year"] 
nojobrate = merged_df["Yearly Unemployment Rate"] 

plt.title('Unemployment Rate %') 

plt.plot(nojobyear, nojobrate,linewidth=2, markersize=12) 
plt.xlabel('Years')
plt.ylabel('Unemployment Rate %')
plt.savefig('linechart')
plt.show()
```

```python
plt.bar(merged_df['Year'],merged_df['Total Restaurant Sales (in millions)'])
plt.xlabel('Years')
plt.ylabel('Total Sales')
plt.title('Restaurant sales (In Millions)')
```

```python
plt.bar(merged_df['Year'],merged_df['Total Vehicle Units (in millions)'],facecolor="red")
plt.xlabel('Years')
plt.ylabel('Units Sold (In Millions)')
plt.title('Vehicle Units Sold (In Millions)')
```
