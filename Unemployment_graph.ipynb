```python
# Import matplotlib for charting
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
# Create a path
unemployment_by_year = Path("unemployment_by_year.csv")
```

```python
# Create DataFrame and plot
unemployment_df = pd.read_csv('unemployment_by_year.csv')
plt.bar(unemployment_df['Year'],unemployment_df['Yearly Unemployment Rate'], color="blue")
plt.xlabel('Years')
plt.ylabel('Unemployment (in Millions)')
plt.title('Unemployment Rate Through the Years')

plt.show()
```

```python
nojobyear = unemployment_df["Year"] 
nojobrate = unemployment_df["Yearly Unemployment Rate"] 

plt.title('Unemployment Rate %') 

plt.plot(nojobyear, nojobrate,linewidth=2, markersize=12) 
plt.xlabel('Years')
plt.ylabel('Unemployment Rate %')
plt.savefig('linechart')
plt.show()
```
